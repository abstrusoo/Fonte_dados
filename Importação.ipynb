```python
origem = '/content/drive/My Drive/DataImp/'
```

```python
arquivo_1 = origem + 'IMP_2025_MUN.csv'
arquivo_2 = origem + 'IMP_2024_MUN.csv'
arquivo_3 = origem + 'IMP_2023_MUN.csv'
arquivo_4 = origem + 'IMP_2022_MUN.csv'
arquivo_5 = origem + 'IMP_2021_MUN.csv'
NCM = origem + 'NCM.csv'
PAIS = origem + 'PAIS.csv'
VIA = origem + 'VIA.csv'
URF = origem + 'URF.csv'
UF_MUN = origem + 'UF_MUN.csv'
```

```python
impPais = pd.read_csv(PAIS,low_memory=False,sep=';',encoding='latin1')
impUfmun = pd.read_csv(UF_MUN,low_memory=False,sep=';',encoding='latin1')
```

```python
imp25mun = pd.read_csv(arquivo_1,low_memory=False,sep=';',encoding='UTF-8')
imp24mun = pd.read_csv(arquivo_2,low_memory=False,sep=';',encoding='UTF-8')
imp23mun = pd.read_csv(arquivo_3,low_memory=False,sep=';',encoding='UTF-8')
imp22mun = pd.read_csv(arquivo_4,low_memory=False,sep=';',encoding='UTF-8')
imp21mun = pd.read_csv(arquivo_5,low_memory=False,sep=';',encoding='UTF-8')
```

```python
Imp_Final = pd.concat([imp24mun,imp25mun,imp23mun,imp22mun,imp21mun], ignore_index=True)
```

```python
Imp_Final = Imp_Final.merge(impPais[['CO_PAIS' , 'NO_PAIS']],on='CO_PAIS',how='left')
```

```python
Imp_Final = Imp_Final.merge(impUfmun[['CO_MUN_GEO' , 'NO_MUN']], left_on='CO_MUN', right_on='CO_MUN_GEO', how='left')
# Drop the redundant CO_MUN_GEO column after merging
Imp_Final = Imp_Final.drop('CO_MUN_GEO', axis=1)
```

```python
Imp_Final_sp = Imp_Final[Imp_Final['SG_UF_MUN'].str.startswith('SP')]
```

```python
municipios_interesse = ['SAO JOSE DOS CAMPOS', 'JACAREI', 'TAUBATE', 'CACAPAVA']
Imp_Final_sp_filtered_mun = Imp_Final_sp[Imp_Final_sp['NO_MUN'].isin(municipios_interesse)]
```

```python
Imp_Final_sp_filtered_mun.to_csv(origem + 'Imp_Final_sp_filtered_mun.csv', index=False)
```
