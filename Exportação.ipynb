```python
origem = '/content/drive/My Drive/Data/'
```

```python
arquivo_5 = origem + 'EXP_2025_MUN.csv'
arquivo_4 = origem + 'EXP_2024_MUN.csv'
arquivo_3 = origem + 'EXP_2023_MUN.csv'
arquivo_2 = origem + 'EXP_2022_MUN.csv'
arquivo_1 = origem + 'EXP_2021_MUN.csv'
UF_MUN = origem + 'UF_MUN.csv'
PAIS = origem + 'PAIS.csv'
```

```python
Exp25mun = pd.read_csv(arquivo_1,low_memory=False,sep=';',encoding='UTF-8')
Exp24mun = pd.read_csv(arquivo_2,low_memory=False,sep=';',encoding='UTF-8')
Exp23mun = pd.read_csv(arquivo_3,low_memory=False,sep=';',encoding='UTF-8')
Exp22mun = pd.read_csv(arquivo_4,low_memory=False,sep=';',encoding='UTF-8')
Exp21mun = pd.read_csv(arquivo_5,low_memory=False,sep=';',encoding='UTF-8')
ExpPais = pd.read_csv(PAIS,low_memory=False,sep=';',encoding='latin1')
ExpUfmun = pd.read_csv(UF_MUN,low_memory=False,sep=';',encoding='latin1')
```

```python
exp_final = pd.concat([Exp24mun,Exp25mun,Exp23mun,Exp22mun,Exp21mun], ignore_index=True)
```

```python
exp_final.info()
```

```python
exp_final = exp_final.merge(ExpPais[['CO_PAIS' , 'NO_PAIS']],on='CO_PAIS',how='left')
```

```python
exp_final.head(1000)
```

```python
exp_final = exp_final.merge(ExpUfmun[['CO_MUN_GEO' , 'NO_MUN']],left_on='CO_MUN', right_on='CO_MUN_GEO',how='left')
```

```python
exp_final.head(1000)
```

```python
exp_final = exp_final[exp_final['SG_UF_MUN'].str.startswith('SP')]
```

```python
exp_final.info()
```

```python
exp_final = exp_final[exp_final['NO_MUN'].isin(['SAO JOSE DOS CAMPOS' , 'JACAREI' , 'TAUBATE' , 'CACAPAVA'])]
```

```python
exp_final.info()
```

```python
exp_final.to_csv(origem + 'Exportação_Dados.csv', index=False)
```
